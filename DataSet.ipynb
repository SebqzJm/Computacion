```python
import pandas as pd
import numpy as np

profesores = {
    "ID": ["001", "002", "003", "004", "005", "006", "007", "008", "009", "010"],
    "Nombre": ["Andres", "Esteban", "Sthephanie", "Sharon", "Alejandro", "Alex", "Katy", "Eduardo", "Jose", "Pancho"],
    "Apellido": ["Altos", "Quiñonez", "Reynosa", "Diaz", "Pineda", "Santos", "Gutierres", "Muñoz", "Ibañez", "Salazar"],
    "Edad": ["33", np.nan, "37", "57", "25", np.nan, "22", "63", "27", "40"],
    "Curso": ["Lenguaje", "Sociales", np.nan, "Matematicas", "Biologia", "Quimica", "Programacion", np.nan, "Seminario", "Anatomia"],
}

registro1 = pd.DataFrame(profesores)

print("--- Máscara de nulos (isnull) ---")
print(registro1.isnull()) 

print("--- Conteo de nulos por columna ---")
print(registro1.isnull().sum())

print("--- ¿Es nulo? (isna) ---")
print(registro1["Curso"].isna().sum())

print("\n--- Filas donde la Edad NO es nula ---")
print(registro1[registro1["Edad"].notnull()])

registro1["Curso"] = registro1["Curso"].fillna("Por asignar")
registro1["Edad"] = registro1["Edad"].fillna("0")

print("\n--- DataFrame Limpio con fillna ---")
print(registro1)
```
